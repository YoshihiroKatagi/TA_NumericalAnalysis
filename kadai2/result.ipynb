```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
#kadai2-1
filename = "result1.csv"
df = pd.read_csv(filename)
```

```python
df
```

```python
plt.plot(df["n"],df["trapezoidal"],label="trapezoidal")
plt.plot(df["n"],df["midpoint"],label="midpoint")
plt.plot(df["n"],df["simpson"],label="simpson")
plt.legend()
plt.show()
```

```python
answer = 31 / math.log(2)
print(answer)
```

```python
# E)
plt.plot(df["n"], abs(answer - df["trapezoidal"]), label="trapezoidal")
plt.plot(df["n"], abs(answer - df["midpoint"]), label="midpoint")
plt.plot(df["n"], abs(answer - df["simpson"]), label="simpson")
plt.title("Absolute Error")
plt.xlabel("fraction size")
plt.ylabel("Value")
plt.legend()
plt.savefig("result1.png")
plt.show()
```
