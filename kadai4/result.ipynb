```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
#絶対誤差
filename = "AbsoluteError.csv"
df = pd.read_csv(filename)
df
```

```python
plt.plot(df["W"], df["trapezoidal"], label="trapezoidal")
plt.plot(df["W"], df["midpoint"], label="midpoint")
plt.plot(df["W"], df["simpson"], label="simpson")
plt.title("Absolute Error")
plt.xlabel("Integral interval")
plt.ylabel("Value")
plt.legend()
plt.savefig("AbsoluteError.png")
plt.show()
```

```python
#相対誤差
filename = "RelativeError.csv"
df = pd.read_csv(filename)
df
```

```python
plt.plot(df["W"], df["trapezoidal"], label="trapezoidal")
plt.plot(df["W"], df["midpoint"], label="midpoint")
plt.plot(df["W"], df["simpson"], label="simpson")
plt.title("Relative Error")
plt.xlabel("Integral interval")
plt.ylabel("Value")
plt.legend()
plt.savefig("RelativeError.png")
plt.show()
```
