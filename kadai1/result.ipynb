```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = "result_changeA1.csv"
df = pd.read_csv(filename)
```

```python
df
```

```python
plt.plot(df["A1"],df["sigma1"],label="Thermal stress 1")
plt.plot(df["A1"],df["sigma2"],label='Thermal stress 2')
plt.legend()
plt.savefig("TS_A1.png")
plt.show()
```

```python
plt.plot(df["A1"],df["delta1"],label="Expand and Contract 1")
plt.plot(df["A1"],df["delta2"],label='Expand and Contract 2')
plt.legend()
plt.savefig("EC_A1.png")
plt.show()
```
