```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
#オイラー法
filename = "Euler_result.csv"
df = pd.read_csv(filename)
df
```

```python
plt.plot(df["t"], df["x"], label="Euler Method")
# plt.plot(df["t"], df["x_true"], label="Analytical Solution")
plt.title("Euler Method")
plt.xlabel("t")
plt.ylabel("x")
plt.legend()
plt.savefig("Euler_t_x.png")
plt.show()
```

```python
plt.plot(df["x"], df["dx"], label="Euler Method")
plt.title("Euler Method")
plt.xlabel("x")
plt.ylabel("dx")
plt.legend()
plt.savefig("Euler_x_dx.png")
plt.show()
```
