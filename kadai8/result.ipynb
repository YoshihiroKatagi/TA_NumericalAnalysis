```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
#オイラー法
filename = "data.csv"
df = pd.read_csv(filename)
df
```

```python
plt.plot(df["t"], df["x"], label="t-x")
plt.title("Graph1")
plt.xlabel("t")
plt.ylabel("x")
plt.legend()
plt.savefig("Graph1-1.png")
plt.show()
```

```python
plt.plot(df["t"], df["v"], label="t-v")
plt.title("Graph2")
plt.xlabel("t")
plt.ylabel("v")
plt.legend()
plt.savefig("Graph1-2.png")
plt.show()
```

```python
plt.plot(df["x"], df["v"], label="x-v")
plt.title("Graph3")
plt.xlabel("x")
plt.ylabel("v")
plt.legend()
plt.savefig("Graph1-3.png")
plt.show()
```
