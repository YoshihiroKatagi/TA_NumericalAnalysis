```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
for i in range(4):
  print("result" + str(i+1))
  filename = "data" + str(i+1) + ".csv"
  df = pd.read_csv(filename)
  print(df)

  print("t-x")
  plt.plot(df["t"], df["x"], label="t-x")
  plt.title("Graph1")
  plt.xlabel("t")
  plt.ylabel("x")
  plt.legend()
  plt.savefig("Graph" + str(i+1) +"-1.png")
  plt.show()

  print("t-v")
  plt.plot(df["t"], df["v"], label="t-v")
  plt.title("Graph2")
  plt.xlabel("t")
  plt.ylabel("v")
  plt.legend()
  plt.savefig("Graph" + str(i+1) +"-2.png")
  plt.show()

  print("x-v")
  plt.plot(df["x"], df["v"], label="x-v")
  plt.title("Graph3")
  plt.xlabel("x")
  plt.ylabel("v")
  plt.legend()
  plt.savefig("Graph" + str(i+1) +"-3.png")
  plt.show()
```
